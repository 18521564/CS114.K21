import thư viện

```python
import turicreate
```

đọc file home_data.sframe lưu vào sales


```python
sales = turicreate.SFrame('/content/drive/My Drive/home_data.sframe/')
```

```python
sales
```

# Khám phá dữ liệu



```python
sales.show()
```

```python
turicreate.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

# Mô hình hồi quy đơn giản từ squarefeet(diện tích)  đoán giá nhà

```python
training_set, test_set = sales.random_split(.8,seed=0)
```

## huấn luyện mô hình

```python
sqft_model = turicreate.linear_regression.create(training_set,target='price',features=['sqft_living'])
```

# Đánh giá chất lượng mô hình thông qua các độ đo như R2 hay RMSE


```python
print (test_set['price'].mean())
```

```python
print (sqft_model.evaluate(test_set))
```

# Có thể khai thác dữ liệu sâu hơn

```python
sqft_model.coefficients
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_set['sqft_living'],test_set['price'],'.',
        test_set['sqft_living'],sqft_model.predict(test_set),'-')
```

# Khám phá nhiều khía cạnh của dữ liệu

# Xây dựng model với các yếu tố này

```python
my_features_model = turicreate.linear_regression.create(training_set,target='price',features=my_features)
```

# So sánh model đơn giản (chỉ squarefeet) với model phúc tạp mới xây dựng xong


```python
print (my_features)
```

```python
print (sqft_model.evaluate(test_set))
print (my_features_model.evaluate(test_set))
```
